```python
TEXT_DATA_PATH = "tiny_shiekspear.txt"
with open(TEXT_DATA_PATH, "r") as file:
    text_data = file.read()
    
print(text_data[:100])  # Print the first 500 characters of the text data
```

```python
import tiktoken
enc = tiktoken.get_encoding("gpt2")
tokens = enc.encode(text_data)
print(tokens[:24])
```

```python
import torch
buf = torch.tensor(tokens[:24 + 1])
x = buf[:-1].view(4,6)
y = buf[1:].view(4,6)

print(x)
print(y)
```
